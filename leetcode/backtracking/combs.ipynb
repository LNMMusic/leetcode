```python
def combs_K_decimals(n: int, k: int, target: int, current: List[int], result: List[List[int]]):
    '''
    - n: from 1 to 9 (must start from 1)
    '''
    # base case
    if k == 0:
        if target == 0:
            result.append(current.copy())
        return

    # limit: range for unique combinations
    # target: avoid opening unnecesary branches
    limit = 11 - k
    while n < limit and n <= target:
        # combs
        # n: increase to avoid repetition
        # k: reduce dimentionality until base case
        # target: we substract the current n
        # current: the involved numbers of the cumulative
        combs_K_decimals(n=n+1, k=k-1, target=target-n, current=current+[n], result=result)
        n += 1
    
    return
```
