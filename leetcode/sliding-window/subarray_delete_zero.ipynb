```python
class Solution:
    def longestSubarray(self, nums: List[int]) -> int:
        length = len(nums)
        # base case
        if length < 2:
            return length == 1
        
        # init
        result = 0
        z = 0
        l = 0
        for r in range(length):
            # increment zero
            if nums[r] == 0:
                z += 1

            # calculate breakpoint
            if z > 1:
                result = max(result, r - 1)
                # shrink l until z back to 0
                while z or l < r:
                    if nums[l] == 0:
                        z -= 1
                    l += 1
                # check if remanent is smaller than result
                # r - l could be 0 (excludes r)
                # length - r includes r
                if result >= (r - l) + (length - r):
                    break
                    
        result = max(result, r - l + 1) # +1 to include r

        return result
```
